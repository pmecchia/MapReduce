```python
import pandas as pd
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
a=pd.read_csv("/Users/pierre/Documents/Telecom/INF727_Systemes_repartis/perf_results/deontologie_police_nationale.csv",sep=",")
a["size_Mo"]=0.007
a
```

```python
b=pd.read_csv("/Users/pierre/Documents/Telecom/INF727_Systemes_repartis/perf_results/amazon_file-1.csv",sep=",")
b["size_Mo"]=398.8
b
```

```python
c=pd.read_csv("/Users/pierre/Documents/Telecom/INF727_Systemes_repartis/perf_results/enwik9.csv",sep=",")
c["size_Mo"]=1024
c
```

```python
all=pd.concat([a,b,c])
all['t_without_network']=all['split_input']+all['map']+all['shuffle']+all['reduce']+all['result']
```

```python
fig = px.line(all, x="nb_machines", y="total",color='size_Mo', markers=True)
fig.update_layout(title='Temps en fonction du nombre de machines pour différentes tailles de fichiers',
                  xaxis_title='Nombre de machines',
                  yaxis_title='Temps en secondes')
fig.show()
```

```python
fig = px.line(all, x="nb_machines", y="t_without_network",color='size_Mo', markers=True)
fig.update_layout(title='Temps (réseau exclu) en fonction du nombre de machines pour différentes tailles de fichiers',
                  xaxis_title='Nombre de machines',
                  yaxis_title='Temps en secondes')
fig.show()
```

```python
fig = make_subplots(rows=4, cols=2,subplot_titles=(steps.columns),x_title='Nombre de machines',
                    y_title='Temps en secondes')


color_dict={'0.007':'#636efa','398.8':'#EF553B','1024.0':'#00cc96'}


for idx in range(steps.shape[1]):
    row=int(np.floor(idx/2)+1)
    col=int(idx%2+1)
    fi=px.line(all,x="nb_machines",y=all.iloc[:,idx],color="size_Mo")
    for d in fi.data:
        fig.add_trace(
            go.Scatter(x=d["x"], y=d['y'],name=d['name']+" Mo",line=dict(color=color_dict[d['name']]),legendgroup=d['name'],showlegend=False if idx!=0 else True),
            row=row, col=col
        )



fig.update_layout(
    autosize=False,
    width=800,
    height=600,
    title='Temps par étapes en fonction du nombre de machines'
    )

fig.show()
```
